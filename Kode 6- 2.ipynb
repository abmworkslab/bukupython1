```python
from scipy import stats

x = df["Hours"]
y = df["Scores"]


slope, intercept, r, p, std_err = stats.linregress(x, y)

def garis(x):
  return slope * x + intercept

model = list(map(garis, x))

plt.scatter(x,y)
plt.plot(x,model)
plt.xlabel("Hours")
plt.ylabel("Scores")
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
X = df["Hours"].values.reshape(-1,1)
y = df["Scores"].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
waktu_budi = 10
prediksi = model.predict([[waktu_budi]])
if prediksi > 100 : 
  prediksi =  100  
else : 
  prediksi =  prediksi 
print("Nilai Budi: ",prediksi)
```

```python
df.drop(columns="Id", inplace = True)
```

```python
jumlah_row = 4
jumlah_column = 2 
lebar = 12
tinggi = 15
fig, axs = plt.subplots(jumlah_row,jumlah_column,figsize=(lebar,tinggi))
sns.histplot(data=df, x="SepalLengthCm",kde=True, ax=axs[0][0])
sns.boxplot(data=df, x="SepalLengthCm", ax=axs[0][1])
sns.histplot(data=df, x="SepalWidthCm",kde=True, ax=axs[1][0])
sns.boxplot(data=df, x="SepalWidthCm", ax=axs[1][1])
sns.histplot(data=df, x="PetalLengthCm",kde=True, ax=axs[2][0])
sns.boxplot(data=df, x="PetalLengthCm", ax=axs[2][1])
sns.histplot(data=df, x="PetalWidthCm",kde=True, ax=axs[3][0])
sns.boxplot(data=df, x="PetalWidthCm", ax=axs[3][1])
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
X = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y = df["Species"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

scaler = StandardScaler()
model = LogisticRegression(solver='lbfgs')
model_final= Pipeline([('standardize', scaler),
                    ('log_reg', model)])
model_final.fit(X_train.values, y_train)
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score

prediksi = model_final.predict(X_test.values)
accuracy_score(y_test,prediksi)
```
